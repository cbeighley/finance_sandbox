# Playing with Quandl data

```python
import quandl
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
t10 = quandl.get('FRED/GS10')
months = 10*12
t10_moving_avg = pd.rolling_mean(t10, months, center=True)
```

```python
fig, ax = plt.subplots(figsize=(16,5))
ax.set_title("10 Year Treasuries Rates, Monthly")
ax.set_ylabel("Interest Rate")
ax = t10.plot(ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(16,5))
ax.set_title("10 Year Treasuries Rates, Monthly, %d month moving average" % months)
ax.set_ylabel("Interest Rate")
ax = t10_moving_avg.plot(ax=ax)
```

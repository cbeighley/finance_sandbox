# Simple plot to show the cost of index fund expense ratios

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def er_cost(years, cagr, expense_ratio):
    return (cagr**years - (cagr-expense_ratio)**years)/(cagr**years)
```

```python
cagr = 1.07
years = 40

expense_ratios = np.array([i*0.0001 for i in range(5,150,5)])
er_costs = np.array([er_cost(years, cagr, i) for i in expense_ratios])
```

```python
plt.plot(expense_ratios*100, er_costs, marker='o', linestyle='None')
plt.xlabel('Expense ratio (% of money in fund per year)')
plt.ylabel('Total %s of fund value paid in expenses by end of %s years' % ('%', years))
plt.title('Cost of expense ratio over a %s year investing period' % (years))
plt.show()
```
